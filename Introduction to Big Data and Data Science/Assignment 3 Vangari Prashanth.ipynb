```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Q1

#1.
'''
Plots the line plot of sin(x) and cos(x + π/2) for -2π < x < 2π.

The sin(x) graph is plotted in red, and the cos(x+π/2) graph is plotted in green.
'''

x = np.linspace(-2*np.pi, 2*np.pi, 17)
s = np.sin(x)
c = np.cos(x + np.pi/2)


plt.figure(figsize=(8, 6))
plt.plot(x, s, 'r', label='sin(x)')
plt.plot(x, c, 'g', label='cos(x+π/2)')
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
#2.

"""
    Plots a subplot with two graphs:
    - sin(x)
    - cos(x+π/2)
"""
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.plot(x, s, 'r')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('sin(x)')

plt.subplot(2, 1, 2)
plt.plot(x, c, 'g')
plt.xlabel('x')
plt.ylabel('cos(x+π/2)')
plt.title('cos(x+π/2)')

plt.tight_layout()
plt.show()
```

```python
#3D Plots

from mpl_toolkits.mplot3d import Axes3D


x = np.arange(-np.pi, np.pi, 0.1)
y = np.arange(-np.pi, np.pi, 0.1)
x, y = np.meshgrid(x, y)
z = np.sin(x**2) + np.cos(y**2)

# Contour plot
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.contour(x, y, z, cmap='viridis')
plt.title('Contour Plot')
plt.xlabel('x')
plt.ylabel('y')


ax = plt.subplot(1, 2, 2, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis', linewidth=0, antialiased=False)
plt.title('Surface Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
